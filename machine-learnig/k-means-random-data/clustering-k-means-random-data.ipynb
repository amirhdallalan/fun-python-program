```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
#set random seed
np.random.seed(0)
```

```python
X, y, cntrs = make_blobs(n_samples=4000, centers=5, n_features=2, cluster_std=1.4, random_state=44, return_centers=True)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
print('Xs :\n', X[0:10])
print('ys :\n', y[0:10])
print('Centers :\n', cntrs)
```

```python
# K-means with n clusters
kmeans = KMeans(n_init = 10)
kmeans.fit(X)
```

```python
kmeans_labels = kmeans.labels_
kmeans_cluster_centers = kmeans.cluster_centers_
```

```python
fig = plt.figure(figsize=(5, 5))
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(kmeans_labels))))
ax = fig.add_subplot(1, 1, 1)
for k, col in zip(range(len(set(kmeans_labels))), colors):
    my_members = kmeans_labels == k
    cluster_center = kmeans_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)
ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())
plt.show()
```

```python
# K-means with 5 clusters
kmeans = KMeans(init = 'k-means++', n_clusters=5, n_init=10)
kmeans.fit(X)
```

```python
kmeans_labels_5 = kmeans.labels_
kmeans_cluster_centers_5 = kmeans.cluster_centers_
```

```python
fig = plt.figure(figsize=(5, 5))
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(kmeans_labels_5))))
ax = fig.add_subplot(1, 1, 1)
for k, col in zip(range(len(set(kmeans_labels_5))), colors):
    my_members = kmeans_labels_5 == k
    cluster_center = kmeans_cluster_centers_5[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)
ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())
plt.show()
```
